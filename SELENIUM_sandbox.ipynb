```python
!pip3 install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

from selenium import webdriver
from selenium.webdriver.common.by import By
import unittest
from selenium.webdriver.chrome.options import Options

class GoogleSearchTest(unittest.TestCase):
    def setUp(self):
        chrome_options = Options()
        chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        self.driver = webdriver.Chrome(options=chrome_options)

    def test_search_in_google(self):
        driver = self.driver
        driver.get("http://www.google.com")
        search_box = driver.find_element(By.NAME, 'q')
        search_box.send_keys('OpenAI')
        search_box.submit()
        assert "OpenAI" in driver.title

    def tearDown(self):
        self.driver.close()

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

```python
# Install the Selenium package for web automation
!pip3 install selenium
# Update the system's package list to ensure the latest versions of packages can be installed
!apt-get update
# Install the Chromium webdriver which is needed for Chrome to work with Selenium
!apt install chromium-chromedriver
# Copy the chromedriver to a directory in the system PATH for easy access
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
# Add the chromedriver path to the system path
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

from selenium import webdriver
from selenium.webdriver.common.by import By
import unittest
from selenium.webdriver.chrome.options import Options

# Define a class for our test, inheriting from unittest.TestCase
class GoogleSearchTest(unittest.TestCase):
    # setUp is called before each test function execution. It sets up the test environment.
    def setUp(self):
        # Options for the Chrome webdriver
        chrome_options = Options()
        # Comment out or remove the --headless option to run Chrome in non-headless mode
        # chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        # Initialize the Chrome driver with the specified options
        self.driver = webdriver.Chrome(options=chrome_options)

    # This is the actual test that is run
    def test_search_in_google(self):
        driver = self.driver
        # Open the Google home page
        driver.get("http://www.google.com")
        # Find the search box element by its name attribute
        search_box = driver.find_element(By.NAME, 'q')
        # Type 'OpenAI' into the search box
        search_box.send_keys('OpenAI')
        # Submit the search form
        search_box.submit()
        # Assert that 'OpenAI' is in the page title
        assert "OpenAI" in driver.title

    # tearDown is called after each test function execution. It cleans up the test environment.
    def tearDown(self):
        # Close the browser window
        self.driver.close()

# This if-statement ensures that the tests are run only when the script is executed directly (not imported)
if __name__ == "__main__":
    # Run the test with unittest's test runner, ignoring the first CLI argument
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
